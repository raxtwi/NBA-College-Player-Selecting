```python
built_dataF = pd.DataFrame({'player name':top5of2021['player_name'], 'Asist/Turnover':top5of2021['ast/tov'], 'Recruit Rank':top5of2021['Rec Rank'], 'Overall Pick':top5of2021['overall pick']})
built_dataF
```

```python
sum_x1square = (built_dataF['Asist/Turnover']**2).sum()
print(sum_x1square)
```

```python
sum_x1_x2cube = ((built_dataF['Recruit Rank']**3)*(built_dataF['Asist/Turnover'])).sum()
sum_x1_x2 = (built_dataF['Recruit Rank']*built_dataF['Asist/Turnover']).sum()
print(sum_x1_x2cube)
```

```python
sum_x2sixth = (built_dataF['Recruit Rank']**6).sum()
sum_x2square = (built_dataF['Recruit Rank']**2).sum()
print(sum_x2sixth)
```

```python
sum_x1y1 = (built_dataF['Asist/Turnover']*built_dataF['Overall Pick']).sum()
print(sum_x1y1)
```

```python
sum_x2cube_y2 = ((built_dataF['Recruit Rank']**3)*built_dataF['Overall Pick']).sum()
sum_x2y2 = (built_dataF['Recruit Rank']*built_dataF['Overall Pick']).sum()
print(sum_x2cube_y2)
```

```python
x = np.array([[sum_x1square, sum_x1_x2cube], [sum_x1_x2cube, sum_x2sixth]])
y = np.array([sum_x1y1, sum_x2cube_y2])
import GaussJordan as gj
b = gj.gauss_jordan(x, y, 2)
b#here , the second correlation coefficient can not be shown properly.
```

```python
#lets use x2 as itself , not its cube
x2 = np.array([[sum_x1square, sum_x1_x2], [sum_x1_x2, sum_x2square]])
y2 = np.array([sum_x1y1, sum_x2y2])

b2 = gj.gauss_jordan(x2,y2)
b2
```

```python
#this means , we have y = a1*x1 + a2x2 built model which is 
#a1 = 4.21, a2 = -0.02, x1 = asist/turnover list, x2 = recruit rank list, y = overall pick
a1 = 4.21
a2 = -0.02
```

```python
players2022_dataF = pd.read_csv("CollegeBasketballPlayers2022.csv")
```

```python
players2022_dataF.drop_duplicates(subset = 'player_name', keep='last', inplace=True)
players2022_dataF = players2022_dataF.reset_index(drop=True)
players2022_dataF
```

```python
testdata = pd.DataFrame({'player_name':players2022_dataF['player_name'], 'ast/tov':players2022_dataF['ast/tov'], 'Rec Rank':players2022_dataF['Rec Rank']})
testdata.dropna(how= 'any', inplace=True)
testdata = testdata.reset_index(drop=True)
testdata['Overall Pick'] = (a1*testdata['ast/tov'])+(a2*testdata['Rec Rank'])
testdata
```

```python
testdata2 = pd.DataFrame()

def is_in_range(x):
    if((x >= 1) & (x <= 5.1)):
        return True

overallpicks = pd.Series(testdata['Overall Pick'])
player2022names = pd.Series(testdata['player_name'])
draft2022 = pd.DataFrame()
    
for pick,name in zip(overallpicks, player2022names):
    if(is_in_range(pick)):
        draft2022 = draft2022.append({'player_name':name, 'pick_number':int(pick)}, ignore_index=True)
draft2022 = draft2022.sort_values('pick_number')
draft2022
```

```python
selectedplayers2022 = pd.DataFrame()
draftnumbers = pd.Series([1,2,3,4,5])
for num in draftnumbers:
    selectedplayers2022 = selectedplayers2022.append(draft2022[draft2022['pick_number'] == num].head(1), ignore_index=True)
selectedplayers2022 #this dataframe shows the first 5 players that will be selected in 2022 according to the regression built.
```
